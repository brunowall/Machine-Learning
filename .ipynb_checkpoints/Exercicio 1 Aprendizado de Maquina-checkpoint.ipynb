### Quest√£o 1

```python
%matplotlib inline
from matplotlib import pyplot as plt

import sklearn
from sklearn import linear_model, metrics
from sklearn import datasets
import numpy as np
```

### Questao 2

```python
X, y = datasets.load_diabetes(return_X_y=True)
```

```python
coluna = []
for i in range(0, len(X[0])) :
    
    for j in range(0,len(X)):
        coluna.append(X[j][i])
    plt.figure(i)
    plt.title("Feature x"+str(i))
    plt.grid()
    plt.scatter(coluna, y)
    del coluna[:]

  
```

#### A)
A terceira fature pois os dados est√£o crescente e est√£o menos dispersos

#### B)
A segunda feature pois dados iguais t√£o gerando relacoes diferentes em Y

### Quest√£o 3

```python
coluna = []
for i in range(0, len(X[0])) :
    
    for j in range(0,len(X)):
        coluna.append(X[j][i])
    print ("Correlacao x"+str(i))    
    print(np.corrcoef(coluna, y)[0, 1])
    del coluna[:]
```

### Quest√£o 4

```python
model = linear_model.LinearRegression()
model.fit(X, y)
```

### Quest√£o 5

#### B0

```python
b0 = model.intercept_
model.intercept_
```

#### B1

```python
b1 = model.coef_
model.coef_
```

#### Formula

```python
for i in range (0, X.shape[1]):
    f = b1[i] * X[i] + b0
    plt.plot(f)
    plt.show()
```

### Quest√£o 6

```python
 y_pred  = model.predict(X)
 print(y_pred)   
```

### Quest√£o 7

```python
mse = metrics.mean_squared_error(y, y_pred)
print (mse)
```

```python
mse_manual = sum((y - y_pred) ** 2) / len(y)
print (mse)
```

### Quest√£o 8

```python
X3 = []
for i in range(0, len(X)) :
    item = [X[i][2]]
    X3.append(item)
   
```

```python
modeloX3 = linear_model.LinearRegression()

modeloX3.fit(X3,y) 
```

```python
print(len(X3))

y_pred2  = modeloX3.predict(X3)
b1 = np.sum((X3 - np.mean(X3)) * (y_pred2 - np.mean(y_pred2))) / np.sum((X3 - np.mean(X3)) ** 2) 
print("B1  Manual")
print(b1)

b0 = np.mean(y_pred2) - b1 * np.mean(X3)

print("B0 Manual")
print (b0)
```

```python
print("B1 auto")
print(str(modeloX3.coef_)+"\n")

print("B0 auto")
print(str(modeloX3.intercept_)+"\n")
```

```python
plt.grid()
plt.scatter(X3, y, color='red')
plt.plot(X3, y_pred2)
```

```python
mse = metrics.mean_squared_error(y, y_pred2)
print (mse)
```
